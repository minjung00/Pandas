````python
url = "https://search.naver.com/search.naver"
params = {
  "query": "뉴진스 다니엘", # 검색어
  "ds":"2022.11.01", # 시작일자
  "de":"2022.12.01", # 종료일자
  "where": "news",
  "sort":2,
  "pd":3,
  "start":1 # 페이지 바꿀 때 쓰는 것
} # 딕셔너리형태로 표현된 param으로 url을 구성할 수 있음 (쿼리스트링)
res = requests.get(url, params)
res.text
````

## HTML
* 브라우저 -> 웹페이지 -> '구조' -> DOM(Document Object Model)
* HTML -> Tag들의 묶음 -> 요소(element)
* 내가 가져오고 싶은 데이터가 있는 '요소'를 특정.결정.
* F12 -> 요소/Elements -> Inspect(조사) / Ctrl + Shift + C
* 긁어오고 싶은 부분을 클릭 -> 오른쪽에 검색이 됨
* 지정이 된 것을 오른쪽 클릭 -> 복사 -> Selector로 복사
```
#sp_nws231 > div > div > div.news_info > div.info_group > a
#sp_nws232 > div > div > div.news_info > div.info_group > a
...
```

````python
# 텍스트로 되어 있는 웹페이지 데이터를 주소값을 활용할 수 있는 구조화된 데이터로 변형
from bs4 import BeautifulSoup # 텍스트 데이터를 html로 구조화해주는 라이브러리

soup = BeautifulSoup(res.text)
soup # 1. requests.get 으로 response를 받아서 soup 구조화
````

````python
# selector -> 화면 상의 주소 를 사용해서 특정한 위치의 데이터(태그)를 불러오는 기능
tags = soup.select('div > div > div.news_info > div.info_group > a')
tags # 2. soup를 통해서 selector를 넣어서 (주소, 규칙..) 필요한 태그(요소)들을 검색
````

````python
# 3. 검색된 태그를 내가 원하는 데이터만 추출
[t.get_text() for t in tags] # tag -> get_text => tag(element) 내부의 텍스트 값만 남겨줌
````

````python
def get_data_from_soup(
    soup, # 1단계 : 특정한 URL에, 쿼리스트링을 통해서 요청 후 응답받은 페이지를 구조화
    selector, # 2단계 : 화면 상 특정한 위치의 요소를 검색
    strategy, # 3단계 : 요소의 어떠한 데이터를 추출할지 방식을 결정
):
    tags = soup.select(selector)
    data = [strategy(t) for t in tags]
    return data
````

````python
data = get_data_from_soup(
    soup,
    'div > div > div.news_info > div.info_group > a',
    # 'div > div > div.news_info > div.info_group > a.info.press', # 1. 선택자를 더욱 좁게
    lambda x: x.get_text()
)
[d for d in data if not d == '네이버뉴스'] # 필요없는 데이터를 나중에 필터링
````

````python
url = "https://search.naver.com/search.naver"
# URL로 반복적으로 데이터 수집
params = {
  "query": "뉴진스", # 검색어
  "ds":"2022.12.01", # 시작일자
  "de":"2022.12.08", # 종료일자
  "where": "news",
  "sort":2,
  "pd":3,
  "start":1 # 페이지 바꿀 때 쓰는 것 <- 딕셔너리는 key 호출에서 업데이트
} # 딕셔너리형태로 표현된 param으로 url을 구성할 수 있음 (쿼리스트링)
def get_news(start, debug=False):
    params['start'] = start # 새로운 페이지 부여
    res = requests.get(url, params) # 반복해서 특정한 URL에 대한 데이터 호출
    soup = BeautifulSoup(res.text, 'lxml') # lxml -> 속도 상승
    media = get_data_from_soup(
        soup,
        # 'div > div > div.news_info > div.info_group > a',
        'div > div > div.news_info > div.info_group > a.info.press',
        lambda x: x.get_text()
    )
    date = get_data_from_soup(
        soup,
        'div > div > div.news_info > div.info_group > span',
        lambda x: x.get_text()
    )
    # '면' 문제 해결
    date = [d for d in date if '면' not in d] # d(date의 원소) 안에 '면'이란 단어가 없는 것만 남기기
    title = get_data_from_soup(
        soup,
        # 'div > div > a',
        'div > div > a.news_tit',
        lambda x: x.get_text()
    )
    link = get_data_from_soup(
        soup,
        # 'div > div > a',
        'div > div > a.news_tit',
        lambda x: x['href']
    )
    if debug:
        print(res.url)
        print(media)
        print(len(media))
        print(date)
        print(len(date))
        print(title)
        print(len(title))
        print(link)
        print(len(link))
    data = pd.DataFrame({
            "날짜": date,
            "언론사": media,
            "제목": title,
            "링크": link,
        })
    return data
````

````python
df = None # df 자리 만들어 놓음
start = 1
while True: # 계속 반복해주면서 데이터를 연결
    try:
        print(f"{start}-{start+9} ({len(df)})")
        news = get_news(start)
        # print('[news]')
        # print(news)
        start += 10
        if (len(news) == 0) or start >= 4000:
            print('반복 종료')
            break
        if df is None:
            df = news
        else:
            df = pd.concat([df, news], ignore_index=True)
        # print('[df]')
        # print(df)
    except Exception as e:
        print(type(e))
        print(e)
        break
df
````

````python
from datetime import datetime, timedelta

now = datetime.utcnow() + timedelta(hours=9)
now
````

````python
def handle_date(x):
    if '전' not in x:
        return x
    dt = datetime.utcnow() + timedelta(hours=9)
    if '시간 전' in x:
        dt -= timedelta(hours=int(x.replace('시간 전', '')))
    if '일 전' in x:
        dt -= timedelta(days=int(x.replace('일 전', '')))
    return dt.date().strftime('%Y.%m.%d.')

df['날짜'] = pd.to_datetime(df['날짜'].apply(handle_date))
df
````
